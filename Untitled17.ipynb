```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB # Naive Bayes juga diimport
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print("Library berhasil diimport.")
```

```python
data = pd.read_csv('iris.csv')
print(data.head(5))
```

```python
drop_column = ["Id"]
data.drop(drop_column, axis=1, inplace=True)
print(data.head())
```

```python
x = data.iloc[:, :-1].values
y = data['Species']
```

```python
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=27)
```

```python
print(X_train) # Untuk menampilkan data training fitur
print(y_train) # Untuk menampilkan data training label
```

```python
SVC_model = svm.SVC()
# Inisialisasi model KNN (dengan 5 neighbors)
KNN_model = KNeighborsClassifier(n_neighbors=5)

# Melatih model (fit) menggunakan data training
SVC_model.fit(X_train, y_train)
KNN_model.fit(X_train, y_train)
```

```python
SVC_prediction = SVC_model.predict(X_test)
KNN_prediction = KNN_model.predict(X_test)
```

```python
# Accuracy score
print(accuracy_score(SVC_prediction, y_test))
print(accuracy_score(KNN_prediction, y_test))

# Confusion Matrix (untuk SVM)
print(confusion_matrix(SVC_prediction, y_test))
print(classification_report(KNN_prediction, y_test))
```
